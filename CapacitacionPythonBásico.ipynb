#### Listas

Estructura de datos que permite almacenar una colección ordenada y mutable de elementos. Se define utilizando corchetes [] y los elementos se separan por comas

**Usos comunes de listas:**

- Almacenar colecciones de elementos: Las listas se utilizan para almacenar colecciones de elementos de cualquier tipo. (Datos con la misma estructura)

- Iteración: Las listas son iterables, lo que significa que puedes recorrer sus elementos utilizando bucles for.

- Almacenar datos estructurados: Las listas pueden contener elementos de diferentes tipos, lo que las hace útiles para almacenar datos estructurados (Tablas, archivos csv, json).

- Implementación de pilas y colas: Las listas se pueden usar para implementar estructuras de datos como pilas y colas.

```python
lista1 = [1, 2, 3, 4, 5]

dato=lista1[0]
print(dato)

datos=lista1[0:2]
print(datos)

dato=lista1[-1]
print(dato)

datos=lista1[2:-1]
print(datos)

datos=lista1[:-2]
print(datos)

# Modificar datos de la lista
lista1[2]=6
print(lista1)

print(lista1[0:])
```

#### Tuplas
Secuencia inmutable de elementos.

**Usos comunes de las tuplas:**

- Retorno múltiple de funciones: Las funciones pueden devolver múltiples valores empacados en una tupla.

- Iteración eficiente: Las tuplas son más eficientes que las listas para la iteración cuando los elementos no necesitan modificarse.

- Uso en diccionarios: Las tuplas pueden usarse como claves en diccionarios de Python si contienen objetos inmutables.

- Datos heterogéneos: Las tuplas pueden contener elementos de diferentes tipos.

```python
# Usando paréntesis
tupla1 = (1, 2, 3, 4, 5)

print(tupla1[0])  # Output: 1
print(tupla1[3])  # Output: 4
```

```python
tupla1[0] = 10  # Esto generará un error
```

```python
# Ejemplo de retorno múltiple de una función
def coordenadas():
    return 10, 20

x, y = coordenadas()
print("Coordenada X:", x)  # Output: 10
print("Coordenada Y:", y)  # Output: 20
```

#### Dicionarios

Es una colección desordenada y mutable de pares clave-valor. Se define utilizando llaves {} y los elementos se especifican en forma de clave: valor.

**Usos comunes de diccionarios:**

- Almacenar datos estructurados: Los diccionarios son útiles para almacenar datos estructurados donde los datos están organizados en pares clave-valor.

- Búsqueda eficiente: Dado que los diccionarios están implementados como tablas hash, proporcionan una búsqueda eficiente de valores basados en sus claves.

- Representar datos JSON: Los diccionarios en Python son similares a los objetos JSON y son útiles para representar datos en formato JSON.

- Mapeo de datos: Los diccionarios se utilizan para mapear un conjunto de claves a un conjunto de valores.

- Configuración de parámetros: Los diccionarios pueden utilizarse para representar configuraciones de parámetros donde cada clave representa un parámetro y su valor representa su configuración.

```python
mi_diccionario = {"nombre": "Juan", "edad": 30, "ciudad": "Madrid"}
print(mi_diccionario)

print(mi_diccionario["nombre"])
print(mi_diccionario["edad"])

# Modeificar diccionario
mi_diccionario["edad"] = 31
print(mi_diccionario)

# Agregar clave
mi_diccionario["profesion"] = "Ingeniero"
print(mi_diccionario)

del mi_diccionario["ciudad"]
print(mi_diccionario)
```

## Funciones



Es un bloque de código reutilizable que realiza una tarea específica cuando es llamado o invocado. Las funciones permiten organizar el código de manera modular, lo que facilita su mantenimiento, reutilización y comprensión.

```python
# Ejemplo función sin return
# Hace las acciones que tenga dentro de la función como imprimir en pantalla o en un archivo externa
def Mensaje(nombre):
    print("¡Hola, " + nombre + "!")

# Llamando a la función
msj='Mi nombre es Juan'
Mensaje(msj)  # Output: ¡Hola, Juan!
```

```python
# Ejemplo función con return
# Hace operaciones
def Mensaje(nombre):
    salida= "¡Hola, " + nombre + "!"
    return salida

# Llamando a la función
msj='Mi nombre es Juan'
varible=Mensaje(msj)

print(varible)
```

```python
def square(x):
  print('La variable inicial es:', x )
  sqr=x**2
  print('La variable al elevada al cuadrado es:', sqr )
  return sqr

result = square(2)
print(result)
```

```python
def square(x,p):
  print('La variable inicial es:', x )
  sqr=x**p
  print('La variable al elevada a la potencia ' + str(p) + ' es:', sqr )
  return sqr

result = square(2,3)
print('Resultado por duera de la función: ', result)
```

### Función Lambda

 lambda es una función anónima y pequeña que se define sin un nombre utilizando la palabra clave lambda. Son útiles para escribir funciones cortas y simples en línea, sin necesidad de definirlas formalmente con def. Las funciones lambda pueden tener cualquier número de argumentos, pero solo pueden tener una expresión. Se utilizan principalmente cuando se necesita una función para una operación rápida y temporal.

```python
# Definir una función lambda que devuelve el cuadrado de su argumento
doble = lambda x: x ** 2

# Llamar a la función lambda
print(doble(2))
```

## Clases

En programación orientada a objetos (OOP), una clase es una plantilla que define las propiedades y comportamientos de un objeto. Una clase puede considerarse como un plano o un molde a partir del cual se pueden crear múltiples instancias u objetos. Cada objeto creado a partir de una clase específica comparte las mismas características (atributos) y comportamientos (métodos), pero puede contener datos diferentes.

image.png

image.png

```python
class Coche:
    def __init__(self, marca, modelo, color):
        self.marca = marca
        self.modelo = modelo
        self.color = color
        self.encendido = False
        self.VelMax=0


    def encender(self):
        if not self.encendido:
            self.encendido = True
            print("El coche ha sido encendido.")
        else:
            print("El coche ya está encendido.")

    def apagar(self):
        if self.encendido:
            self.encendido = False
            print("El coche ha sido apagado.")
        else:
            print("El coche ya está apagado.")

    def VelObj(self,VelInt):

        if VelInt>self.VelMax:
          msj='No puede alcanzar la velidad objetivo'
          print(msj)
          return msj


# Crear una instancia de la clase Coche
mi_coche = Coche("Toyota", "Corolla", "Rojo")

# Acceder a los atributos y llamar a los métodos de la instancia
print(f"Mi coche es un {mi_coche.marca} {mi_coche.modelo} de color {mi_coche.color}.")
print(mi_coche.encendido)
mi_coche.encender()
mi_coche.encender()
mi_coche.apagar()
mi_coche.encender()
```

## Librerías


Las librerías de Python son colecciones de funciones y módulos predefinidos que se pueden utilizar para llevar a cabo una variedad de tareas específicas. Estas librerías están diseñadas para ampliar las capacidades del lenguaje Python, proporcionando funcionalidades adicionales y herramientas para diferentes propósitos, como ciencia de datos, desarrollo web, procesamiento de texto, análisis numérico, entre otros.

Algunas librerías son:

- **NumPy**: Es una librería fundamental para computación científica en Python. Proporciona soporte para matrices y operaciones matemáticas de alto nivel, lo que lo hace esencial para el análisis numérico y la manipulación de datos.

- **Pandas**: Es una librería de análisis de datos que proporciona estructuras de datos y herramientas para trabajar con datos estructurados, como tablas y series temporales. Es ampliamente utilizada en ciencia de datos y análisis de datos.

- **Matplotlib**: Es una librería de trazado que se utiliza para crear visualizaciones de datos en 2D, como gráficos de líneas, histogramas y diagramas de dispersión. Es esencial para la visualización de datos en Python.

- **Requests**: Es una librería HTTP para Python que permite realizar solicitudes HTTP de manera sencilla y eficiente. Es utilizada para interactuar con servicios web y realizar solicitudes a APIs.

- **Django y Flask**: Son dos frameworks web populares en Python. Django es un framework de desarrollo web de alto nivel que facilita la construcción rápida de aplicaciones web completas, mientras que Flask es un framework más ligero y flexible que se utiliza para construir aplicaciones web más pequeñas y simples.

Se utiliza las figuras que son objetos que tienen atributos y métodosp para ajustar las gráficas de acuerdo al requerimiento del usuario.


```python
# Gráficos usando figuras
import numpy as np
import matplotlib.pyplot as plt

# La figura crea un espacio donde dibujar el gráfico
fig = plt.figure()

# Añadimos los límites para crear un objeto de ejes donde dibujar
rect = (0, 0, 1, 1)
axes = fig.add_axes(rect)

# En los ejes se crea el gráfico mediante plt
axes.plot(np.random.randint(100,size=[6]), label="Pedro")
axes.plot(np.random.randint(100,size=[6]), label="Marta")
axes.plot(np.random.randint(100,size=[6]), label="Ana")

# El mapeado de nombres ahora requiere usar dos métodos diferentes
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
mapeado = range(len(meses))
axes.set_xticks(mapeado)
axes.set_xticklabels(meses)
```

```python
fig.set_size_inches(2, 2)
fig.set_dpi(100)
fig
```

## Ficheros de texto

Exiten varias funciones para lectura de arhivos, las siguiente son usando la liberería io

#### Escritura

```python
#creación y escritura de un fichero
from io import open

texto = "Línea 1 \n Línea 2"

#Ruta donde se va a crear el fichero
fichero = open('fichero.txt','w')

#Se escribe el texto
fichero.write(texto)

#Se cierra el fichero
fichero.close()
```

#### Lectura

```python
#Lectura de un fichero

from io import open

#Ruta donde se va a leer el fichero
fichero = open('fichero.txt','r')

#escribimo el texto
texto = fichero.readlines()

#Cerramos el fichero
fichero.close()
print(texto)
```

## Dataframe


Un DataFrame es una estructura de datos tabular bidimensional y mutable en la biblioteca de análisis de datos pandas de Python. Se puede considerar similar a una hoja de cálculo o una tabla de base de datos SQL, donde los datos se organizan en filas y columnas etiquetadas.

```python
import pandas as pd
import numpy as np

array = np.random.uniform(-10, 10, size=[4,4])

df = pd.DataFrame(array, index=['A','B','C','D'], columns=['W','X','Y','Z'])

df
```
